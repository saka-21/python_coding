```python
# 装飾
x = np.linspace(-10, 10, 10)
z = 1 / (1 + np.exp(-x))
```

```python
plt.plot(x, z,
         color='green',
         linewidth=2,
         ls='--',
         marker='o',
         markersize=10,
         markerfacecolor='r',
         markeredgecolor='b',
         markeredgewidth=4,
         alpha=0.4)
```

```python
# 付属情報をつける
plt.plot(x, z, label='sigmoid')
plt.plot(x, z*2, label='sigmoid * 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sigmoid(x)')
plt.legend(loc=4)
plt.xticks([1, 2, 3], rotation=20)
plt.yticks([0, 0.5, 1], rotation=20)
```

```python
# csvファイルの読み込み
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
# release_dateがNaNのレコードをフィルタアウト
df = df[~df['release_date'].isna()]
# yearカラムを新規作成
df['year'] = df['release_date'].apply(lambda x: int(x[:4]))
```

```python
df
```

```python
plt.plot(df.groupby('year').count()['id'])
plt.xlabel('year')
plt.ylabel('count')
```

```python
# revenueの推移
df = df[df['revenue']!=0]
plt.plot(df.groupby('year')['revenue'].median(), label='median')
plt.plot(df.groupby('year')['revenue'].mean(), label='mean')
plt.plot(df.groupby('year')['revenue'].max(), label='max')
plt.legend()
```

## plt.fugure()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
x = np.linspace(-10, 10, 100)
ax1.plot(x, x**2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, x**2)
axes[1].plot(x, x**3)
```

```python
fig, axes = plt.subplots(2, 2)
axes[0,0].plot(x, x**2)
axes[0,1].plot(x, x**3)
axes[1,0].plot(x, x**4)
axes[1,1].plot(x, x**5)
fig.tight_layout()
```

```python
fig, axes = plt.subplots()
axes.plot(x, x**2)
```

```python
fig, axes = plt.subplots()
axes.plot(x, x**2)
plt.xlabel('x')
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, x**2, label='x**2')
axes[1].plot(x, x**3)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('title')
axes[0].set_xticks([-3, 1, 5])
axes[0].legend()
axes[1].axis('off')
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
df.head()
```

```python
plt.scatter(df['revenue'], df['budget'], alpha=0.2)
```

```python
df = df[(df['revenue'] != 0) & (df['budget'] != 0)]
# df.groupby('original_language')
lang_group_df = df.groupby('original_language').mean()[['budget', 'revenue', 'vote_count']]
c = np.arange(len(lang_group_df))
plt.scatter('revenue', 'budget', s='vote_count', c=c, data=lang_group_df)
for idx, row in lang_group_df.iterrows():
    plt.annotate(idx, (row['revenue'], row['budget']))
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
plt.hist(df['vote_average'], bins=50)
plt.show()
```

```python
## profit（revenue - budget）のヒストグラムを書いてみよう
fig, axes = plt.subplots()
df = df[(df['revenue'] != 0) & (df['budget'] != 0)]
df['profit'] = df['revenue'] - df['budget']
logbins = np.logspace(5, np.log10(df['profit'].max()), 100)
axes.hist(df['profit'], bins=logbins)
axes.set_xscale('log')
```

### plotを保存する

```python
x = np.linspace(-3, 3, 10)
```

```python
plt.figure(figsize=(10, 6), dpi=100)
plt.plot(x, x**2)
```

```python
fig, axes = plt.subplots(figsize=(10, 5))
axes.plot(x, x*2)
```

```python
## pngファイルで保存: .savefig()
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
plt.scatter('runtime', 'budget', data=df, alpha=0.2)
plt.xlabel('runtime')
plt.ylabel('budget')
plt.savefig('runtime_budget_scatter')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 3))
x = np.linspace(-3, 3, 100)
axes[0].set_title('cosine')
axes[0].plot(x, np.cos(x))
axes[1].set_title('sine')
axes[1].plot(x, np.cos(x))
fig.tight_layout()
fig.savefig('cos_sin.png')
```

```python
## pdfで保存
from matplotlib.backends.backend_pdf import PdfPages

pdf = PdfPages('pdf_files.pdf')

fig, axes = plt.subplots()
axes.plot(x, x**2)
pdf.savefig(fig)
pdf.close()
```

```python
pdf = PdfPages('savemultifig_sample.pdf')
for i in range(10):
    fig, axes = plt.subplots()
    axes.plot(x, x**i)
    axes.set_title('x^{}'.format(i))
    pdf.savefig(fig)
pdf.close()
```

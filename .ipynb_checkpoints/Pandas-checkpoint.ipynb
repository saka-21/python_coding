```python
import pandas as pd
import numpy as np
```

```python
# dictionayからSeriesを作成
data = {
    'Name': 'John',
    'Sex': 'male',
    'Age': 22
}
```

```python
pd.Series(data)
```

```python
# Numpy ArrayからSeriesを作成
array = np.array([22, 31, 42, 23])
age_series = pd.Series(array)
```

```python
array = np.array(['John', 'male', 22])
john_series = pd.Series(array, index=['Name', 'Sex', 'Age'])
```

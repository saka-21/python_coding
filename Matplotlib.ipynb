```python
# 装飾
x = np.linspace(-10, 10, 10)
z = 1 / (1 + np.exp(-x))
```

```python
plt.plot(x, z,
         color='green',
         linewidth=2,
         ls='--',
         marker='o',
         markersize=10,
         markerfacecolor='r',
         markeredgecolor='b',
         markeredgewidth=4,
         alpha=0.4)
```

```python
# 付属情報をつける
plt.plot(x, z, label='sigmoid')
plt.plot(x, z*2, label='sigmoid * 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sigmoid(x)')
plt.legend(loc=4)
plt.xticks([1, 2, 3], rotation=20)
plt.yticks([0, 0.5, 1], rotation=20)
```

```python
# csvファイルの読み込み
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
# release_dateがNaNのレコードをフィルタアウト
df = df[~df['release_date'].isna()]
# yearカラムを新規作成
df['year'] = df['release_date'].apply(lambda x: int(x[:4]))
```

```python
df
```

```python
plt.plot(df.groupby('year').count()['id'])
plt.xlabel('year')
plt.ylabel('count')
```

```python
# revenueの推移
df = df[df['revenue']!=0]
plt.plot(df.groupby('year')['revenue'].median(), label='median')
plt.plot(df.groupby('year')['revenue'].mean(), label='mean')
plt.plot(df.groupby('year')['revenue'].max(), label='max')
plt.legend()
```

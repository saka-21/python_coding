```python
# 装飾
x = np.linspace(-10, 10, 10)
z = 1 / (1 + np.exp(-x))
```

```python
plt.plot(x, z,
         color='green',
         linewidth=2,
         ls='--',
         marker='o',
         markersize=10,
         markerfacecolor='r',
         markeredgecolor='b',
         markeredgewidth=4,
         alpha=0.4)
```

```python
# 付属情報をつける
plt.plot(x, z, label='sigmoid')
plt.plot(x, z*2, label='sigmoid * 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sigmoid(x)')
plt.legend(loc=4)
plt.xticks([1, 2, 3], rotation=20)
plt.yticks([0, 0.5, 1], rotation=20)
```

```python
df[filter_series]
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
df[(df['original_language'] == 'ja') & (df['vote_average'] > 8)]
```

```python
df.describe()
```

```python
# 予算(budget)が0もしくは売り上げ(revenue)が0の映画
```

```python
df[(df['budget'] == 0) | (df['revenue'] == 0)]
```

```python
# 予算も売り上げも0ではない
```

```python
# df[(df['budget'] != 0) & (df['revenue'] != 0)]
df[~((df['budget'] == 0) | (df['revenue'] == 0))]
```
